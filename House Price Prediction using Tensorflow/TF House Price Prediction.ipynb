```python
df=df.drop('zipcode',axis=1) #we can also make this a categorize feature with expensive zipcodes vs less expensive but for that we should have domain knowledge
#and creating dummy variables will be so much as we have 70 diff zipcodes so here we can drop this feature.
```

```python
df=df.drop('date',axis=1)
```

```python
X=df.drop('price',axis=1).values
y=df['price'].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=101,test_size=0.3)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler=MinMaxScaler()
```

```python
X_train=scaler.fit_transform(X_train)
```

```python
X_test=scaler.fit_transform(X_test)
```

```python
from tensorflow.keras.models import Sequential
```

```python
from tensorflow.keras.layers import Dense
```

```python
model=Sequential()

model.add(Dense(19,activation='relu'))#typically we use no of neurons = no of features
model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu')) 
model.add(Dense(19,activation='relu'))

model.add(Dense(1)) #final layer
```

```python
model.compile(optimizer='adam',loss='mse') #adam is the mostly used and effective optimizer
```

```python
model.fit(x=X_train,y=y_train,validation_data=(X_test,y_test),epochs=400,batch_size=128) #typically batch_size is in power of 2
#validation data will not effect the values of weights and biases, this is only used for testing purpose
```

```python
pred_df=pd.DataFrame(model.history.history)
```

```python
pred_df.head()
```

```python
 pred_df.plot() #as we can see both val_loss and loss are decreasing that means there is no overfitting and we can also continue our training to get more better results
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,explained_variance_score
```

```python
y_pred=model.predict(X_test)
```

```python
np.sqrt(mean_squared_error(y_test,y_pred))
```
