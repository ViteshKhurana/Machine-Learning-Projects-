```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv('datasets/cancer_classification.csv')
```

```python
X=df.drop('benign_0__mal_1',axis=1).values
y=df['benign_0__mal_1'].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=101,test_size=0.25)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler=MinMaxScaler()
```

```python
X_train=scaler.fit_transform(X_train)
```

```python
X_test=scaler.transform(X_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Dropout
```

```python
model1=Sequential()

model1.add(Dense(30,activation='relu'))

model1.add(Dense(15,activation='relu'))

model1.add(Dense(1,activation='sigmoid')) #we use sigmoid as our activation function in last layer because our problem is binary classification and this function will output between 0 and 1

model1.compile(optimizer='adam',loss='binary_crossentropy') #for binary classification task
```

```python
model1.fit(X_train,y_train,epochs=600,validation_data=(X_test,y_test))
```

```python
model_history1=pd.DataFrame(model1.history.history)
model_history1
```

```python
model_history1.plot() #here we can see that the val_loss is increasing after a certain point. This is the condition of overfitting
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
y_pred1=model1.predict_classes(X_test)
```

```python
print(classification_report(y_test,y_pred1))
print(confusion_matrix(y_test,y_pred1))
```

# Early stopping for improving model 

```python
model2=Sequential()

model2.add(Dense(30,activation='relu'))

model2.add(Dense(15,activation='relu'))

model2.add(Dense(1,activation='sigmoid')) 

model2.compile(optimizer='adam',loss='binary_crossentropy')
```

```python
from tensorflow.keras.callbacks import EarlyStopping
```

```python
early_stopping=EarlyStopping(monitor='val_loss',patience=25,mode='min',verbose=1) #patience is the no of epochs for which model
#will continue training after an earlystopping is occured. This will help us to check if the early stopping is because of 
#noise or model is overfitting after that point.
```

```python
model2.fit(X_train,y_train,epochs=600,validation_data=(X_test,y_test),callbacks=[early_stopping])
```

```python
model_history2=pd.DataFrame(model2.history.history)
```

```python
model_history2.plot()
```

```python
y_pred2=model2.predict_classes(X_test)
```

```python
print(classification_report(y_test,y_pred2))
print(confusion_matrix(y_test,y_pred2))
```

```python
model3=Sequential()

model3.add(Dense(30,activation='relu')) #dropout will randomly deactivate some of the neurons in the layer
model3.add(Dropout(0.5)) #this will take rate between 0 to 1 that indicates what percentage of neurons we want to deactivate

model3.add(Dense(15,activation='relu'))
model3.add(Dropout(0.5))

model3.add(Dense(1,activation='sigmoid')) 

model3.compile(optimizer='adam',loss='binary_crossentropy')
```

```python
model3.fit(X_train,y_train,epochs=600,validation_data=(X_test,y_test),callbacks=[early_stopping])
```

```python
model_history3=pd.DataFrame(model3.history.history)
```

```python
model_history3.plot()
```

```python
y_pred3=model3.predict_classes(X_test)
```

```python
print(classification_report(y_test,y_pred3))
print(confusion_matrix(y_test,y_pred3))
```
